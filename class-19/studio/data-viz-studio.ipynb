```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to 
## get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", on_bad_lines='skip')

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(int(date[2]))

df["year"] = years

df.head()
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.Total books released between 2001 to 2010.
#year = np.sort(df[df['year'] > 2000]['year'].unique())
year = np.sort(df[(df['year'] > 2000) & (df['year'] < 2011)]['year'].unique())

print(year)
number_of_books = []
for y in year:
	sum = df[df["year"] == y]['year'].count()
	number_of_books.append(sum)
print(number_of_books)
plt.bar(x=year, height=number_of_books,width=.8, color ='blue')

plt.title("Total books released every year between 2001 to 2010")
plt.xlabel("Years")
plt.ylabel("Total Books")

```

```python
# Use your Python skills to create a visualization here.
# Single line chart
years = np.sort(df[df['year'] > 2000]['year'].unique())
years_df = df[df['year'] > 2000]
avg_ratings_year = np.sort(years_df.groupby("year")["average_rating"].mean())


plt.xticks(np.arange(2000, 2024, 2))
plt.plot(years, avg_ratings_year)
plt.title("Average rating since year 2000 ")
plt.xlabel("Years")
plt.ylabel("Average rating")


```
